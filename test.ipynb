```python
import yfinance as yf
import pandas as pd
from datetime import date, timedelta
import stocks

def get_shares_outstanding(ticker_symbol):
    """Obtient le nombre d'actions en circulation pour un symbole boursier."""
    try:
        ticker = yf.Ticker(ticker_symbol)
        info = ticker.info
        if "sharesOutstanding" in info:
            return info["sharesOutstanding"]
        else:
            return None
    except Exception as e:
        print(f"Erreur lors de la récupération des informations pour {ticker_symbol}: {e}")
        return None

def calculate_market_cap_change(ticker_symbol):
    """Calcule la variation de la capitalisation boursière pour un symbole boursier."""
    try:
        today = date.today()
        yesterday = today - timedelta(days=1)
        data = yf.download(ticker_symbol, start=yesterday, end=today)

        if data.empty:
            return None

        shares_outstanding = get_shares_outstanding(ticker_symbol)
        if not shares_outstanding:
            return None

        price_yesterday = data["Close"].iloc[0]
        price_today = data["Close"].iloc[-1]

        market_cap_yesterday = price_yesterday * shares_outstanding
        market_cap_today = price_today * shares_outstanding

        return market_cap_today - market_cap_yesterday

    except Exception as e:
        return None

def find_top_market_cap_changes(tickers, top_n=10):
    """Trouve les n actions avec les plus grandes variations de capitalisation boursière."""
    changes = {}
    for ticker in tickers:
        change = calculate_market_cap_change(ticker)
        if change is not None:
            changes[ticker] = change

    sorted_changes = sorted(changes.items(), key=lambda item: item[1], reverse=True)
    return sorted_changes[:top_n]

# Indices à analyser
indices = {
    "S&P 500": stocks.SP500_TICKERS,
    "CAC 40": stocks.CAC40_TICKERS
}
```

```python
# Nombre d'actions à afficher
top_n = 5

for index_name, tickers in indices.items():
    print(f"\nTop {top_n} variations de capitalisation boursière pour {index_name}:")
    top_changes = find_top_market_cap_changes(tickers, top_n)
    if top_changes:
        for ticker, change in top_changes:
            print(f"{ticker}: {change:.2f}")
    else:
        print("Impossible de récupérer les données.")
```
